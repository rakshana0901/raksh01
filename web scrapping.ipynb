```python
from bs4 import BeautifulSoup
```

```python
import requests
```

```python
url="https://www.imdb.com/list/ls062622141/"
```

```python
res=requests.get(url)
```

```python
res
```

```python
doc=BeautifulSoup(res.text)
```

```python
doc
```

```python
name=doc.select(".lister-item-header a")
```

```python
name
```

```python
year=doc.select(".text-muted.unbold")
```

```python
year
```

```python
ratings=doc.select(".ipl-rating-star.small .ipl-rating-star__rating")
```

```python
ratings
```

```python
genre=doc.select(".genre")
```

```python
genre
```

```python
synopsis=doc.select(".ipl-rating-widget+ p , .ratings-metascore+ p")
```

```python
synopsis
```
